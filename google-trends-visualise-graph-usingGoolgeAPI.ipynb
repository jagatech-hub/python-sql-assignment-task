```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import pandas as pd

# Connect to Google Trends API
pytrends = TrendReq()

# Define search terms for 3 terms: Football, Rugby and Tennis and time period 3 Month
search_terms = ['Football', 'Rugby', 'Tennis']
timeframe = 'today 3-m'  # Past 3 months

# Get Google Trends data
pytrends.build_payload(kw_list=search_terms, timeframe=timeframe)
trends_data = pytrends.interest_over_time()

# Plotting
plt.figure(figsize=(10, 6))
for term in search_terms:
    plt.plot(trends_data.index, trends_data[term], label=term)

plt.title('Share of Search Over the Past 3 Months')
plt.xlabel('Date')
plt.ylabel('Share of Search (%)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
