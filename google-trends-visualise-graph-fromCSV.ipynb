```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the 3 Months (Wordlwide) search data for 3 terms: Football, Rugby and Tennis CSV file into a DataFrame
# This trent data also viewed in below Google Trends URL
# https://trends.google.com/trends/explore?date=today%203-m&q=Football,Rugby,%2Fm%2F07bs0&hl=en-US
df = pd.read_csv("C:\\Users\Admin\\Downloads\\multiTimeline_FRT_Worldwide.csv")

# Convert the 'Day' column to datetime format
df['Day'] = pd.to_datetime(df['Day'], format='%d/%m/%Y')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Day'], df['Football:'], label='Football', marker='o')
plt.plot(df['Day'], df['Rugby:'], label='Rugby', marker='o')
plt.plot(df['Day'], df['Tennis:'], label='Tennis', marker='o')
plt.title('Share of Football, Rugby and Tennis Search Over the Past 3 Months')
plt.xlabel('Date')
plt.ylabel('Share of Search (%)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
